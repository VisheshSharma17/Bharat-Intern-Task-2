```python
#RENAMING COLUMNS
data=data.rename(columns={'Revenue (Millions)':'Revenue.Millions'})
data=data.rename(columns={'Runtime (Minutes)':'Runtime.Minutes'})
```

```python
#SETS A FILTER TO IGNORE WARNINGS GENERATED BY SYNTAX
import warnings
warnings.filterwarnings("ignore")
```

```python
#REMOVE ROWS WITH MISSING VALUES FROM DATAFRAME
data= data.dropna()
```

```python
#NO MISSING VALUES LEFT
data.isnull().sum()
```

```python
#DROPPING A COLUMN
data=data.drop("Genre", axis=1)
```

```python
#CREATING HEATMAP OF CORRELATION MATRIX FOR THE DATAFRAME
plt.subplots(figsize=(30,30))
sns.heatmap(data.corr(), annot= True, fmt='.2f')
```

```python
#TOP 10 DIRECTORS OF MOVIES
data.Director.value_counts()[:10].plot.pie(autopct='%1.1f%%',figsize=(10,10))
plt.title("TOP 10 DIRECTORS OF MOVIES")
plt.show()
```

```python
#TOP 10 ACTORS OF MOVIES
data.Actors.value_counts()[:10].plot.pie(autopct='%1.1f%%',figsize=(10,10))
plt.title("TOP 10 ACTORS OF MOVIES")
plt.show()
```

```python
#RATING ON YEAR BASIS 
sns.stripplot(x="Year", y="Rating", data=data, jitter=True);
plt.title("RATING BASED ON YEAR")
plt.show()
```

```python
#REVENUE ON YEAR BASIS 
sns.stripplot(x="Year", y="Revenue.Millions", data=data, jitter=True,color="violet");
plt.title("REVENUE BASED ON YEAR")
plt.show()
```

```python
#METASCORE BASED ON YEAR BASIS 
sns.stripplot(x="Year", y="Metascore", data=data, jitter=True, color="green");
plt.title("METASCORE BASED ON YEAR")
plt.show()
```

```python
#METASCORE BOXPLOT
metascore = data.Metascore
sns.boxplot(metascore,color = "red");
plt.show()
```
